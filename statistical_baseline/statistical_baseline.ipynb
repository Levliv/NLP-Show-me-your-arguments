```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.head()
```

```python
values = train.label.value_counts().sort_values(ascending=False)
values
```

`Zero` is most common, so we will make our best guess with zero on test dataset

```python
def count_accuracy(correct, predictions):
    if correct.shape[0] != predictions.shape[0]: # np arrays must have same sizes
        return None
    counter = 0
    for correct, prediction in zip(correct, predictions):
        counter += int(correct == prediction)
    return counter / predictions.shape[0]
```

```python
print(count_accuracy(np.array([0] * test.label.shape[0]), test.label))
```

### This way we've shown that making our best guess we can only count on 60% accuracy.
